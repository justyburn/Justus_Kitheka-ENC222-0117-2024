```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, first_name, initial, last_name):
        self.first_name = first_name
        self.initial = initial
        self.last_name = last_name

    @abstractmethod
    def calculate_wages(self):
        pass

    def display_details(self):
        print(f"Name: {self.first_name} {self.initial}. {self.last_name} | Wages: ${self.calculate_wages()}")

class SalaryEmployee(Employee):
    def __init__(self, fn, i, ln, salary):
        super().__init__(fn, i, ln)
        self.salary = salary
    
    def calculate_wages(self):
        return self.salary

class HourlyEmployee(Employee):
    def __init__(self, fn, i, ln, hours, rate):
        super().__init__(fn, i, ln)
        self.hours = hours
        self.rate = rate
    
    def calculate_wages(self):
        return self.hours * self.rate

# Output processing
employees = [
    SalaryEmployee("John", "K", "Surveyor", 4500.00),
    HourlyEmployee("Jane", "A", "Analyst", 160.0, 25.0)
]

for emp in employees:
    emp.display_details()
```
