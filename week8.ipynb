```python
class Square:
    def __init__(self, side):
        self.side = side
    def get_peri(self): return 4 * self.side
    def get_area(self): return self.side ** 2

class Cube(Square):
    def __init__(self, side):
        super().__init__(side)
    def get_area(self): return 6 * (self.side ** 2)
    def get_volume(self): return self.side ** 3

monument = Cube(5.0)
print(f"Surface Area: {monument.get_area()}")
print(f"Volume: {monument.get_volume()}")
```
