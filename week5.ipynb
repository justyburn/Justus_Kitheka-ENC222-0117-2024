```python
def get_max_bearing(data, k1, k2, k3):
    # Simulating reference selection via dictionary keys
    max_k = max([k1, k2, k3], key=lambda k: data[k])
    return max_k

bearings = {"b1": 45.5, "b2": 120.3, "b3": 89.1}
key = get_max_bearing(bearings, "b1", "b2", "b3")
print(f"Max: {bearings[key]}")
```
